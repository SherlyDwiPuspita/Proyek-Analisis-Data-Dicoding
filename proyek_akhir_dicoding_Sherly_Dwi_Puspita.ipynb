# Visualization & Explanatory Analysis

### Pertanyaan ke-1 : Bagaimana pola penggunaan sepeda pada hari libur dibandingkan dengan hari kerja?

```python
# Membuat grup berdasarkan hari libur dan hari kerja
day_df['holiday'] = day_df['holiday'].astype('category')
day_df['workingday'] = day_df['workingday'].astype('category')

# Membuat plot untuk membandingkan jumlah pengguna sepeda pada hari libur dan hari kerja
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
sns.lineplot(x='dateday', y='count', hue='holiday', data=day_df, palette=['#FFD700', '#000000'])
plt.title('Pola Penggunaan Sepeda pada Hari Libur dan Hari Kerja', fontsize=18)
plt.xlabel('Tanggal', fontsize=14)
plt.ylabel('Jumlah Pengguna Sepeda', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Hari Libur/Hari Kerja', fontsize=12)
plt.show()
```

### Pertanyaan ke-2 : Bagaimana cuaca mempengaruhi jumlah penyewaan sepeda?

```python
# Membuat grup berdasarkan cuaca
day_df['weathersit'] = day_df['weathersit'].astype('category')

# Menghitung rata-rata jumlah pengguna sepeda untuk setiap kategori cuaca
weathersit_mean = day_df.groupby('weathersit')['count'].mean().reset_index()

# Membuat plot bar chart
plt.figure(figsize=(9, 8))
sns.barplot(x='weathersit', y='count', data=weathersit_mean, color='#FFD700')  # Mengatur warna bar menjadi kuning
plt.title('Pengaruh Cuaca terhadap Jumlah Penyewaan Sepeda', fontsize=16)
plt.xlabel('Cuaca', fontsize=14)
plt.ylabel('Jumlah Pengguna Sepeda', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### Pertanyaan ke-3 : Bagaimana fluktuasi jumlah penyewaan sepeda dari tahun ke tahun, dan apa yang dapat kita simpulkan tentang perubahan perilaku pengguna?

```python
# Mengelompokkan data untuk mendapatkan jumlah penyewaan per bulan dan tahun
monthly_rentals_data = day_df.groupby(['year', 'month'])['count'].sum().reset_index()
monthly_rentals_data['month'] = pd.Categorical(monthly_rentals_data['month'], categories=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], ordered=True)

# Mengambil tahun yang unik untuk membuat subplot
unique_years = monthly_rentals_data['year'].unique()
total_years = len(unique_years)

# Membuat subplot untuk setiap tahun
fig, axes = plt.subplots(total_years, 1, figsize=(10, 6), sharex=True)

# Menentukan warna untuk setiap tahun
colors = ['#FFD700', '#000000']  # Kuning dan Hitam

# Looping untuk setiap tahun
for i, year in enumerate(unique_years):
    yearly_data = monthly_rentals_data[monthly_rentals_data['year'] == year]

    # Mengatur warna untuk setiap tahun, bergantian antara kuning dan hitam
    sns.lineplot(ax=axes[i], x='month', y='count', data=yearly_data, marker='o', color=colors[i % len(colors)])

    # Menambahkan judul dan label
    axes[i].set_title(f'Jumlah Penyewaan Sepeda di Tahun {year}', fontsize=14)
    axes[i].set_ylabel('Jumlah Penyewaan', fontsize=12)
    axes[i].grid(True)

# Menambahkan label untuk sumbu x di subplot terakhir
axes[-1].set_xlabel('Bulan', fontsize=14)

# Mengatur layout agar lebih rapi
plt.tight_layout()

```

# Analisis Lanjutan dengan RFM Analysis

```python
# Melakukan Teknik Anaisis RFM
# Tanggal referensi adalah hari terakhir dalam dataset
reference_date = day_df['dateday'].max()

# Menghitung Recency: jumlah hari sejak sepeda terakhir kali digunakan
day_df['Recency'] = (reference_date - day_df['dateday']).dt.days

# Menghitung Frequency: seberapa sering sepeda digunakan (berdasarkan total pengguna `count`)
frequency_df = day_df.groupby('dateday')['count'].sum().reset_index()
frequency_df.columns = ['dateday', 'Frequency']

# Menghitung Monetary: total pengguna terdaftar (`registered` diasumsikan sebagai pengguna yang membayar)
monetary_df = day_df.groupby('dateday')['registered'].sum().reset_index()
monetary_df.columns = ['dateday', 'Monetary']

# Menggabungkan metrik RFM ke dalam satu DataFrame
rfm_df = day_df[['dateday', 'Recency']].merge(frequency_df, on='dateday').merge(monetary_df, on='dateday')

# Menampilkan DataFrame RFM
print(rfm_df.head())
```

```python
# Asumsikan rfm_df adalah DataFrame Anda
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

# Mendefinisikan warna untuk diagram garis
colors = ["#F57F20", "#F57F20", "#F57F20", "#F57F20", "#F57F20"]

# Plot untuk Recency
sns.lineplot(x="dateday", y="Recency", data=rfm_df.sort_values(by="Recency", ascending=True).head(5), marker='o', color=colors[0], ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Berdasarkan Recency (hari)", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelsize=15)
ax[0].tick_params(axis='x', rotation=45)  # Rotasi label sumbu x

# Plot untuk Frequency
sns.lineplot(x="dateday", y="Frequency", data=rfm_df.sort_values(by="Frequency", ascending=False).head(5), marker='o', color=colors[0], ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("Berdasarkan Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
ax[1].tick_params(axis='x', rotation=45)  # Rotasi label sumbu x

# Plot untuk Monetary
sns.lineplot(x="dateday", y="Monetary", data=rfm_df.sort_values(by="Monetary", ascending=False).head(5), marker='o', color=colors[0], ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("Berdasarkan Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
ax[2].tick_params(axis='x', rotation=45)  # Rotasi label sumbu x

# Menambahkan judul utama untuk semua plot
plt.suptitle("Hari Terbaik Berdasarkan Parameter RFM (dateday)", fontsize=20)

# Menampilkan plot
plt.show()
```
